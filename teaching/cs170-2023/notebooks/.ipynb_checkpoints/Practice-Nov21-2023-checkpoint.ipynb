### Problem 1
Consider Ackley Function [link](https://en.wikipedia.org/wiki/Ackley_function) on the interval $[-5; 5] \times [-5; 5]$:
$$f(x,y) = -20\exp\left[-0.2\sqrt{0.5(x^2+y^2)}\right] -\exp\left[0.5(\cos 2\pi x + \cos 2\pi y))\right] + e + 20$$
Find the global minimum value of this function using Simulated Annealing algorithm with Gaussian state change (finding next point by adding Gaussian random values). Plot the progressive steps on the contour map of the function.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(x, y):
    res = np.exp(1) + 20
    res += -20*np.exp(-0.2*np.sqrt((x**2 + y**2)/2))
    res += - np.exp((np.cos(2*np.pi*x) + np.cos(2*np.pi*y))/2)
    return res
```

```python
x_base = np.linspace(-5, 5, 100)
y_base = np.linspace(-5, 5, 100)

xx, yy = np.meshgrid(x_base, y_base, sparse=True)

z = f(xx, yy)
plt.contourf(x_base, y_base, z)
plt.show()
```

```python
x, y = np.random.uniform(-5, 5), np.random.uniform(-5, 5)
E = f(x, y)

trajectory = [[x,y]]

sigma=1
T = 10
```

```python
while T >= 1e-5:
    for i in range(100):
        x_new = np.random.normal(x, sigma)
        y_new = np.random.normal(y, sigma)
        
        E_new = f(x_new, y_new)
        dE = E_new - E
        if dE < 0:
            x, y = x_new, y_new
        else:
            if np.random.random() < np.exp(-dE/T):
                x, y = x_new, y_new
        trajectory.append([x, y])
    T = T/10
    sigma = sigma / 2
    
trajectory = np.array(trajectory)
```

```python
z = f(xx, yy)
plt.contourf(x_base, y_base, z)
plt.plot(trajectory[:,0], trajectory[:,1], 'r.-')
plt.show()
```

```python
plt.plot(np.arange(trajectory.shape[0]), f(trajectory[:,0], trajectory[:,1]), 'b.-')
plt.show()
```
